```python
import pandas as pd #importacao do pandas, para realizar a leitura do CSV
df = pd.read_csv('swd2023.csv')
user_id = df['UserId'].tolist()
#criar a URL para obter dados da API
sdw2023_url = 'https://sdw-2023-prd.up.railway.app/users/'
import requests #realizar chamadas https
import json #manipular dados de retorno da requisicao

def get_user(id):
    #response = requests.get(f'{sdw2023_url}/users/{id}')#passando a ur
    response = requests.get(f'{sdw2023_url}{id}')#passando a ur
    return response.json() if response.status_code == 200 else None

user = [user for id in user_id if(user := get_user(id)) is not None]
print(json.dumps(user,indent=2))


#enriquecimento dos dados

#instalar a biblioteca da OPENAI
#https://platform.openai.com/docs/api-reference/introduction

#chave para acessar a api
#https://platform.openai.com/account/api-keys

openai_api_key = "?"

#implementacao GPT-4 
import openai

openai.api_key = openai_api_key

#https://platform.openai.com/docs/api-reference/chat/create?lang=python - integracao com o CHAT
def generate_ai_news(user):
    completion = openai.ChatCompletion.create(
    model="gpt-3.5-turbo-instruct",
    messages=[
        {
            "role": "system", 
            "content": "Você é um especialista em markting bancário."
         },
        {
            "role": "user", 
            "content": f"Crie uma mensagem para {user['name']} sobre a importância sobre os investimentos (máximo de 100 caracteres)"
        }
    ]
    )
    return completion.choices[0].message,content.strip('\"')#acessando a resposta, strip-> remover as "" da resposta
#carregar lista de usuarios em funcao


for usr  in user:
    news = generate_ai_news(usr)
    print(news)
    user['news'].append({
        ##"icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
         "description": news
    })
    
```
